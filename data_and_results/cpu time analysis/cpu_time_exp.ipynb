```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.dpi'] = 300
```

```python
df_list = [None] * 10
for i in range(10):
    fnm = f'./HPC_runs/CopiedResults/cpu_exp{i+1}.csv'
    #fnm = f'./VM_runs/CopiedResults/cpu_exp{i+1}.csv'
    df_list[i] = pd.read_csv(fnm)
#print(df_list)
df_hpc = pd.concat(df_list, ignore_index= True)

df_list = [None] * 10
for i in range(10):
    #fnm = f'./HPC_runs/CopiedResults/cpu_exp{i+1}.csv'
    fnm = f'./VM_runs/CopiedResults/cpu_exp{i+1}.csv'
    df_list[i] = pd.read_csv(fnm)
#print(df_list)
df_vm = pd.concat(df_list, ignore_index= True)

x = df_hpc['Tmax'].to_list()
y = df_hpc['run_t'].to_list()
x2 = df_vm['Tmax'].to_list()
y2 = df_vm['run_t'].to_list()

plt.plot(x,y, '.')
plt.plot(x2,y2, '.')

#for i in range(len(x)):
#    if x[i] > 90000 and y[i] < 20:
#        print (i, x[i], y[i], df_hpc['train_ctr'].to_list()[i])
```

```python
x = df_hpc['train_ctr']
x2 = df_vm['train_ctr']
plt.plot(x,y,'o', alpha=0.3, color='blue', label='HPC')
plt.plot(x2,y2,'o', alpha=0.3, color='red', label='Windows VM')
plt.legend()
plt.xlabel('Number of Completed Train Processes')
plt.ylabel('Simulation Run Time (s)')
```
